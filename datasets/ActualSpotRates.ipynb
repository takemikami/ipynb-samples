# データセット: Actual Spot Rates

入手元:
https://ide-research.net/publications/data-sets/

入手元からspot_exrates.datを取得し、同じフォルダに置いて実行します.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('spot_exrates.dat')
feature_names = [
  'AUD', # Australian Dollar
  'BEF', # Belgian Franc
  'CAD', # Canadian Dollar
  'FRF', # French Franc
  'DEM', # German Mark
  'JPY', # Japanese Yen
  'NLG', # Dutch Guilder
  'NZD', # New Zealand Dollar
  'ESP', # Spanish Peseta
  'SEK', # Swedish Krone
  'CHF', # Swiss Franc
  'GBP', # UK Pound
]

# データを表示
fig, ax = plt.subplots(nrows = 1, sharex = True)
ax.plot(data, alpha=0.5)
ax.legend(feature_names, loc='upper left', bbox_to_anchor=(1, 1))
ax.grid()
```
