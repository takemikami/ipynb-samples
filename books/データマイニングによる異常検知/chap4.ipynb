# 4. 変化点検知

```python
# ライブラリの準備
import numpy as np
import matplotlib.pyplot as plt
import changefinder
```

## 4.3 変化点検出エンジンChangeFinder

```python
# 変化点のあるデータを作る
data = np.concatenate(
    [np.random.normal(0.7, 0.05, 300),
     np.random.normal(1.2, 0.05, 300),
     np.random.normal(1.4, 0.05, 30),
     np.random.normal(1.6, 0.05, 30),
     np.random.normal(1.8, 0.05, 30),
     np.random.normal(2.0, 0.05, 300)])
data

# データを表示
fig, ax = plt.subplots(nrows = 1, sharex = True)
ax.plot(data, alpha=0.5)
ax.grid()
```

```python
# changefinderによる変化検知
cf = changefinder.ChangeFinder(r=0.01, order=1, smooth=7) # r: 忘却率, order: 自己回帰窓枠サイズ, smooth: 移動平均窓枠サイズ
result = np.empty(len(data))
for i, d in enumerate(data):
    result[i] = cf.update(d)

# データを表示(単一通貨)
fig, ax = plt.subplots(nrows = 2, sharex = True)
ax[0].plot(data, alpha=0.5)
ax[0].grid()
ax[1].plot(result)
ax[1].set_ylim(0,50)
ax[1].grid()
```
